```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
#  visualizing some of the data
%matplotlib inline
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
# checking the file format
digits.data[:5]
```

```python
# creation of a pandas df for this dataset
df = pd.DataFrame(digits.data)
df.head()
```

```python
#appending the target variables in the df
df['target'] = digits.target
df.head()
```

```python
# train split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis='columns'), df.target, test_size=0.2)
```

```python
len(X_test)
```

```python
# Chosing the classifier to train the model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train) #the number of stimators is the number of trees
# The score of the model depends on the number of trees as well as the
```

```python
model.score(X_test, y_test)
```

```python
# building a confussion matrix to see where the model is predicting poorly
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
