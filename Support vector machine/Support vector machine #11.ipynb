Support vector machine: is an ML algorithm used for classification when we have high dimensionality problems. The classification can be binary or multiclass

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
#Appending the target variable in the df
df['target'] = iris.target
df.head()
```

```python
# To check what means each value of the target
iris.target_names
```

```python
# Some explorations to the data
df[df.target==1]
```

```python
#Adding the color target name
df['flower_name']= df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
# Some data visualizaiton
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
#Separate the three spcies in three df
df0= df[df.target==0]
df1= df[df.target==1]
df2= df[df.target==2]
```

```python
#scatterplot
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='red', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='green', marker='+')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='pink', marker='+')
```

```python
#scatterplot
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='black', marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='orange', marker='+')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='purple', marker='+')
```

```python
# Trainning the model using alll the features not only sepal and petal
from sklearn.model_selection import train_test_split
```

```python
#Removing some columns from our df
X = df.drop(['target','flower_name'], axis='columns')
X.head()
```

```python
y = df.target
y
```

```python
# Train the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
