A decision tree is gonna be used when the relationship between the data

For instance in a dataset you want to know if the salary of the employees is more than 100k per year. And you have three features to consider: degree, company and position. If you can not apply a Logistic regression, you will have to build a decision tree to know (based on the features) if the employee will earn more than 100k or not.


How do you know how to select or give priority to the features? If for instance you split by company, and the result in one division is facebook and abc pharma, you will check how many positive responses you will have in each division, if you have too many or all positives you will be having low entropy in the data and high information gain and viceversa if you have high entropy you will be be having low information gain.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('salaries.csv')
df
```

```python
# Divide the df in target and independent variable
inputs = df.drop('salary_more_then_100k',axis='columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
# ML understand numbers not labels, that's why we have to change the numbers of the first 3 columns using ENCODERS
from sklearn.preprocessing import LabelEncoder
```

```python
# Creation of the 3 objects of the class because we have three columns with labels to convert to numbers
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
# In the input df create one more column
inputs['company_n']= le_company.fit_transform(inputs['company'])
inputs['job_n']= le_job.fit_transform(inputs['job'])
inputs['degree_n']= le_degree.fit_transform(inputs['degree'])
inputs.head()
```

```python
# Dropping the label columns
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
# Ready to train th classifier
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
# Now that my model is ready, we can use it to predict
model.score(inputs_n, target)
```
