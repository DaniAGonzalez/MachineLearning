PCA is a technique that reduces the features to Principal Components while maintaining the maximum variance. It might drop the accuracy but also helps to deal with the curse of dimensionality

```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
dataset = load_digits()
dataset.keys()
```

```python
dataset.data.shape
```

```python
#To convert 1D array to 2d array
dataset.data[0].reshape(8,8)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.gray()
plt.matshow(dataset.data[0].reshape(8,8))
```

```python
#To know the unique values
import numpy as np
np.unique(dataset.target)
```

```python
#creation of the df
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
df.head()
```

```python
df.describe()
```

```python
X = df
y = dataset.target
```

```python
#Scale feature before ML modell
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#train the model
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=30)
```

```python
# Creation of the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
#Lets do PCA
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape #this 29 numbers means that it will create 29 new columns explaining/retain the 95% of the most important variation/information
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python

X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
pca = PCA(n_components=2) #selecting 2 components
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
